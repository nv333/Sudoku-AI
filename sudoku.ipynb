You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import numpy as np

def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.
    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.
    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.
    """
    solved_sudoku = SudokuGrid(sudoku)
    not_solved = np.full(shape=(9, 9), fill_value=-1, dtype=int) 
    
    # Check grid is consistent
    if not solved_sudoku.is_valid_grid():  
        return not_solved
    
    # Check grid is solved
    if solved_sudoku.is_done():  
        return solved_sudoku.final_entries  
    
    # Initialise grid with all possible numbers
    solved_sudoku.initial_entries()  
    
    # Solve grid using depth-first search
    solved_sudoku = perform_search(solved_sudoku)  
    # here for dfs()
    
    # If unsolvable
    if not solved_sudoku:  
        return not_solved

    # Return solved sudoku
    return solved_sudoku.final_entries  


class SudokuGrid:
    
    def __init__(self, final_entries):
        
        self.final_entries = final_entries  # Final grid numbers
        self.cell_domain = np.empty(shape=(9, 9), dtype=list)  # Domain for each cell
        
    def is_valid_grid(self):
        valid = False
        
        # Check for duplicates
        for (row, col), entry in np.ndenumerate(self.final_entries):
            if self.is_valid_entry(row, col, entry) == True:  
                valid = True
              
        return valid
        
    def is_done(self):
        done = False
        
        # Check for empty cells
        if 0 not in self.final_entries:  
            done = True
        return done
        
    def is_consistent(self):
        consistent = True
        
        # Check grid is solvable under constraints
        for cell in np.ndindex(9, 9):
            row, col = cell[0], cell[1]
            if self.final_entries[row][col] == 0:
                if len(self.cell_domain[row][col]) < 1:  # Check each cell has at least 1 possible value
                    consistent = False
        return consistent        

    def initial_entries(self):
        # Initialise domains for each cell
        for (row, col), this_entry in np.ndenumerate(self.final_entries):
            self.cell_domain[row][col] = []  
            if this_entry == 0:  # If cell currently empty
                for entry in range(1, 10):  # Check new entry is valid
                    if self.is_valid_entry(row, col, entry):
                        self.cell_domain[row][col].append(entry)  # Add entry to corresponding cell's domain

    def is_valid_entry(self, this_row, this_col, entry):
        valid = True
        
        # Check along rows and columns:
        for other_cell in range(9):
            if other_cell != this_row or other_cell != this_col:  # Check not considering position with itself
                if entry == self.final_entries[other_cell][this_col]:  # If duplicate
                    valid = False
                if entry == self.final_entries[this_row][other_cell]:
                    valid = False
         
        # Check 3x3 square:
        square_hor = this_row - (this_row % 3)
        square_ver = this_col - (this_col % 3)
       
        for row in range(3):
            for col in range(3):
                if (square_hor + row) != this_row and (square_ver + col) != this_col:  # Check considering within square
                    if entry == self.final_entries[square_hor + row][square_ver + col]:  # If duplicate
                        valid = False
                        
        if entry == 0:  # If empty cell
            valid = True  
    
        return valid

    def get_deduced_entry(self):
        # Get cells with only one remaining possible number
        deduced_entry = []
        for row in range(9):
            for col in range(9):
                # If empty and len(domain) is 1
                if len(self.cell_domain[row][col]) == 1 and self.final_entries[row][col] == 0: 
                    deduced_entry.append((row, col))
        return deduced_entry

    def update_domains(self, this_row, this_col, entry):
        # Update cell domains after assignment
        
        # For rows and columns, remove assigned entry from domains
        for i in range(9):
            if entry in self.cell_domain[i][this_col]:
                self.cell_domain[i][this_col].remove(entry)
            if entry in self.cell_domain[this_row][i]:  
                self.cell_domain[this_row][i].remove(entry)
            

        # For squares, " "
        square_hor = this_row - (this_row % 3)
        square_ver = this_col - (this_col % 3)
        for row in range(3):
            for col in range(3):
                if entry in self.cell_domain[square_hor + row][square_ver + col]:
                    self.cell_domain[square_hor + row][square_ver + col].remove(entry)

    def copy_grid(self):
        # grid_copy = sudoku_grid.copy() here
        # Create copy of grid
        grid_copy = SudokuGrid(np.ndarray.copy(self.final_entries))  # Copy final entries
        for (row, col), entries in np.ndenumerate(self.cell_domain):
            grid_copy.cell_domain[row][col] = entries[:]  # Copy every cell domain
        return grid_copy

    def create_updated_grid(self, row, col, entry):
        # Create updated grid after entry assignment with updated domains
        
        # From copy, update grid
        updated_grid = self.copy_grid()
        updated_grid.final_entries[row][col] = entry # Assign entry
        updated_grid.cell_domain[row][col] = []  # Clear domain once entry has been assigned
        updated_grid.update_domains(row, col, entry)  # Update domains

        # For cells with only one possible number left
        deduced_entry = updated_grid.get_deduced_entry()  # Find cell positions
        while len(deduced_entry) > 0:
            row, col = deduced_entry.pop()  # Get cell position

            updated_grid.final_entries[row][col] = updated_grid.cell_domain[row][col][0]  # Assign entry
            updated_grid.cell_domain[row][col] = []  # Clear domain once entry has been assigned
            updated_grid.update_domains(row, col, updated_grid.final_entries[row][col])  # Update domains
            deduced_entry = updated_grid.get_deduced_entry()  # Continue until list exhausted

        return updated_grid  
```

```python
def perform_search(sudoku_grid):
    # Depth-first search
    
    # Pick position for next move
    row, col = choose_next_cell(sudoku_grid)  
    
    # For each possible entry
    entries = sudoku_grid.cell_domain[row][col]
    for entry in entries:
        # Create grid with updated assignments
        next_grid = sudoku_grid.create_updated_grid(row, col, entry)  
        if next_grid.is_done():  # If found solution
            return next_grid  
        # Else, backtrack, continu
        if next_grid.is_consistent():
            searched = perform_search(next_grid)
            if searched != None and searched.is_done():  # If solvable and solved
                return searched  
            
def choose_next_cell(sudoku_grid): 
    # Most constrained variable heuristic:
    most_constrained_cells = get_most_constrained_cells(sudoku_grid)

    if len(most_constrained_cells) == 1:  # If no need to tie-break:
        return most_constrained_cells[0][0], most_constrained_cells[0][1] # Choose based on mcv
    else:
        # Degree heuristic:
        highest_row, highest_col, highest_degree = -1, -1, 0
        for cell in most_constrained_cells:
            this_degree = get_empty_cells(sudoku_grid, cell[0], cell[1])  # Get cell's degree
            if this_degree > highest_degree:  # Favour highest degree
                highest_degree = this_degree
                highest_row, highest_col = cell  
        return highest_row, highest_col  # Choose based on degree

def get_most_constrained_cells(sudoku_grid):
    positions = {}  # For each possible number, initialise cell positions
    for key in range(10):  
        positions[key] = []

    for (row, col), final_entries in np.ndenumerate(sudoku_grid.final_entries):
        if final_entries == 0:  
            positions[len(sudoku_grid.cell_domain[row][col])].append((row, col)) # Store position if cell still empty

    # Find positions of most constrained cells
    for position in range(10):
        if len(positions[position]) > 0:
            return positions[position]

def get_empty_cells(sudoku_grid, row, col):
    empties_count = 0  # Number of empty positions in same row, column and block connected to postion
    # Check row and column
    for this in range(9):
        if sudoku_grid.final_entries[row][this] == 0 or sudoku_grid.final_entries[this][col] == 0:
            empties_count += 1

    # Check square:
    square_row = row - (row % 3)
    square_col = col - (col % 3)
    
    for i_row in range(3):
        for i_col in range(3):
            if sudoku_grid.final_entries[i_row + square_row][i_col + square_col] == 0:
                empties_count += 1

    return empties_count
```
